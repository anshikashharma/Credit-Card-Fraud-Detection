```python
legit=df[df["Class"]==0]
```

```python
# Under sampling 
```

```python
# building a sample dataset containing similar distrubution of legit and fraud transactions 
```

```python
# no. of fraud transactions ---492
```

```python
legit_sample=legit.sample(n=492)
```

```python
legit_sample
```

```python
# Concatienating two dataframes 
```

```python
dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
dataset
```

```python
dataset.groupby('Class').mean()
```

```python
# both the datasets have kinda similar values we got a good sample 
```

```python
# TRain test split 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=dataset.iloc[:,:-1]
```

```python
y=dataset.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression 
model=LogisticRegression()
```

```python
parameters={'penalty':['l2',"l1", 'elasticnet'],
'dual':[False],
'C':[1.0], 
'solver':['lbfgs','liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
'max_iter':[100,200], 
'multi_class':['auto', 'ovr', 'multinomial']}
```

```python
from sklearn.model_selection import GridSearchCV
cv=GridSearchCV(model,parameters,scoring="accuracy",cv=5)
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
cv.fit(x_train,y_train)
```

```python
y_pred=cv.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
